#### Independencia de los errores
$\textbf{La autocorrelacin}$ de una serie de datos discreta de un proceso $X_t$ no es ms
que el coficiente de correlacin de dicho proceso con una versin desplazada de la propia
serie Sea $e_1,...e_n$ una serie de datos el coeficiente de autocorrelacin es:

$$
    \rho_k=\dfrac{\sum_{i=1}^{n-k}(e_i-\bar{e})(e_{i-k}-\bar{e})}{\sum_{i=1}^{n-k}(e_i-\bar{e})^2}
$$
Si $\rho_k$ se aleja del valor 0 hay evidencia de que cada k observaciones hay un patrn de los
residuales pues dichas observaciones estn muy correlacionadas lo que implicar dependencia
de los residuales. Entoncesla idea es contrastar $\rho_k=0$ vs $\rho_k\neq 0$ . R tiene programada la prueba en la funcin $acf$.
En la siguiente grfica se espera que todas las correlaciones estn dentro de la banda de confianza, salvo por la primera que toma el valor de 1. Si para alguna observacin se sale de la banda de confianza se dice que se encontro una correlacin estadsticamente significativa y hay un patrn en los residuales.

$$

#### Multicolinearidad 
En los supuestos de la regresin se pidioque la matriz $X$ fuese de rango completo por lo que debemos verificar que no existan columnas que sean combinaciones lineales de otras. 
Verificar la multicolineridad no es facil a simple vista sin embargo existen formas de detectarla.
1. Una forma muy simple para detectar multicolinearidad es inspeccionar la matriz de correlaciones entre las variables explicativas. ( cor(X) , pairs(X))
2. Hacer regresiones lineales de Xq con el resto de las variables y obtener el coeficiente de determinacin de dicha regresin $R_q^2$ , un coeficiente de deteminacin cercana a 1 nos dice que hay un ajuste muy bueno por lo que se puede decir que existe apfroximadamente una combinacin lineal de estas variables. 
Una forma de obtener indices es calcular lo que se denomina el V IFq (Factor de Inflacin de la varianza) definido como:
$$
VIF_q=\dfrac{1}{1-R_q^2}\hspace{1cm}q\in{0,..,k}
$$
Entonces si $R_q^2\approx 1$ eso se traduce en un $VIF_q$ grande, en la prctica se suele tomar como punto de corte 5 o 10, es decir obtener un $VIFq > 10$ nos habla de que la variable Xq puede ser expresada aproximadamente como combinacin lineal de las restantes y por tanto podemos tener problemas nmericos en las estimaciones.

```python
VIF = pd.DataFrame()
VIF["Covariables"] = x.columns
VIF["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
VIF
```
